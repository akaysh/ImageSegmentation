## K-means

```julia
using ImageSegmentation, Images
fl = load("flower.jpg")
fl = convert(Array{ColorTypes.RGB{FixedPointNumbers.Normed{UInt8,8}},2},fl)
```

```julia
fuzzy_fl = kmeans(fl,3)
```

```julia
fl_matrix = reshape(fuzzy_fl.assignments,size(fl))
```

```julia
map(i->RGB(fuzzy_fl.centers[1,i],fuzzy_fl.centers[2,i],fuzzy_fl.centers[3,i]), fl_matrix)
```

```julia
ep = convert(Array{ColorTypes.RGB{FixedPointNumbers.Normed{UInt8,8}},2},ep)
```

```julia
fuzzy_ep = kmeans(ep,4)
```

```julia
ep_matrix = reshape(fuzzy_ep.assignments,size(ep))
```

```julia
map(i->RGB(fuzzy_ep.centers[1,i],fuzzy_ep.centers[2,i],fuzzy_ep.centers[3,i]), ep_matrix)
```
